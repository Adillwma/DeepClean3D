```python
import math

def angle_between_vectors(vector_a, vector_b):
    # Calculate the dot product of the two vectors
    dot_product = sum(a * b for a, b in zip(vector_a, vector_b))
    
    # Calculate the magnitudes (lengths) of the vectors
    magnitude_a = math.sqrt(sum(a ** 2 for a in vector_a))
    magnitude_b = math.sqrt(sum(b ** 2 for b in vector_b))
    
    # Calculate the angle in radians
    angle_radians = math.acos(dot_product / (magnitude_a * magnitude_b))
    
    # Convert the angle to degrees
    angle_degrees = math.degrees(angle_radians)
    
    return angle_degrees

# Example usage:
vector_a = [0, 1, 0]
vector_b = [0.5, 0.5, 0]
angle = angle_between_vectors(vector_a, vector_b)
print(f"The angle between vector_a and vector_b is {angle} degrees.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np

def rotate_vectors_to_reference(vectors, target_vector, reference_vector): # target vector and reference vector need to be swapped in order here and then swapped in name throughout the function
    # Ensure that vectors, reference_vector, and target_vector are NumPy arrays
    vectors = np.array(vectors)
    reference_vector = np.array(reference_vector)
    target_vector = np.array(target_vector)

    # Calculate the rotation matrix using the cross product
    cross_product = np.cross(target_vector, reference_vector)
    dot_product = np.dot(target_vector, reference_vector)

    # Create the rotation matrix
    skew_matrix = np.array([[0, -cross_product[2], cross_product[1]],
                             [cross_product[2], 0, -cross_product[0]],
                             [-cross_product[1], cross_product[0], 0]])

    rotation_matrix = np.eye(3) + skew_matrix + \
                     np.dot(skew_matrix, skew_matrix) * (1 - dot_product) / np.linalg.norm(cross_product) ** 2

    # Apply the rotation to all vectors
    rotated_vectors = np.dot(rotation_matrix, vectors.T).T

    return rotated_vectors




# Define the parameters of the cone
reference_vector = np.array([0, 0, 1])  # Reference vector
common_start_point = np.array([0, 0, 0])  # Common starting point
target_vector = np.array([1, 0, 0])
num_vectors = 1000  # Number of vectors to generate
cone_angle_degrees = 30  # Angle of the cone in degrees
cone_thickness = 1
max_magnitude = 1.0  # Maximum magnitude of the vectors

# Convert the cone angle from degrees to radians
cone_angle_radians = np.radians(cone_angle_degrees)
cone_thickness_radians = np.radians(cone_thickness)

# Generate random vectors forming a cone
random_vectors = []
for _ in range(num_vectors):
    # Generate random values for spherical coordinates
    theta = np.random.uniform(0, 2 * np.pi)  # Azimuthal angle
    phi = np.random.uniform(cone_angle_radians, cone_angle_radians + cone_thickness_radians)  # Polar angle within the cone #NOTE: Cone thickness is currently [(cone), (cone + cone_thickness)] rather than [(cone - cone_thickness/2), (cone + cone_thickness/2)] 

    # Calculate the Cartesian coordinates
    x = max_magnitude * np.sin(phi) * np.cos(theta)
    y = max_magnitude * np.sin(phi) * np.sin(theta)
    z = max_magnitude * np.cos(phi)

    # Create the random vector
    random_vector = np.array([x, y, z])

    # Ensure that the vector starts from the common starting point
    random_vector += common_start_point

    random_vectors.append(random_vector)


    #angle = angle_between_vectors(reference_vector, random_vector)
    #print(f"The angle between vector_a and vector_b is {angle} degrees.")


# Example usage:
random_vectors = rotate_vectors_to_reference(random_vectors, reference_vector, target_vector)


for vector in random_vectors:
    angle = angle_between_vectors(target_vector, vector)
    print(f"The angle between vector_a and vector_b is {angle} degrees.")




# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the reference vector
ref_vis = reference_vector*10
ax.plot([-ref_vis[0], ref_vis[0]], [-ref_vis[1], ref_vis[1]], [-ref_vis[2], ref_vis[2]], color='r', linewidth=2, label='Reference Vector')


# Plot the random vectors
for vector in random_vectors:
    x, y, z = vector
    #ax.quiver(common_start_point[0], common_start_point[1], common_start_point[2],x, y, z, color='b', arrow_length_ratio=0.1)
    #plot scatter
    ax.scatter(x, y, z, color='b', s=1)

# Plot target vector

targ_vis = target_vector*10
ax.plot([-targ_vis[0], targ_vis[0]], [-targ_vis[1], targ_vis[1]], [-targ_vis[2], targ_vis[2]], color='g', linewidth=2, label='Target Vector')
# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set plot limits
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])

# set view angle
ax.view_init(azim=30, elev=30)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the coordinates of the four corners of the square (A, B, C, and D)
A = np.array([1, 1, 2])
B = np.array([-1, 1, 2])
C = np.array([-1, -1, 1])
D = np.array([1, -1, 1])

# Normalize vector AB to get the x-axis vector U
U = (B - A) / np.linalg.norm(B - A)

# Normalise vector AD to get the y-axis vector V
V = (D - A) / np.linalg.norm(D - A)

# Calculate the normal vector N by taking the cross product of AB and AD
N = np.cross(B - A, D - A)


# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the square
corners = np.array([A, B, C, D, A])
ax.plot(corners[:,0], corners[:,1], corners[:,2], 'bo-')

# Plot the normal vector N
ax.quiver(*A, *N, color='r', label='Normal Vector (N)')

# Plot the x-axis vector U
ax.quiver(*A, *U, color='g', label='X-axis Vector (U)')

# Plot the y-axis vector V
ax.quiver(*A, *V, color='b', label='Y-axis Vector (V)')

# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set axis limits
ax.set_xlim([min(corners[:,0]) - 1, max(corners[:,0]) + 1])
ax.set_ylim([min(corners[:,1]) - 1, max(corners[:,1]) + 1])
ax.set_zlim([min(corners[:,2]) - 1, max(corners[:,2]) + 1])

# Add legend
ax.legend()

# Show the plot
plt.show()
```

```python
import math

def shortest_distance_to_circle(x1, y1, h, k, r):
    distance = abs(math.sqrt((x1 - h) ** 2 + (y1 - k) ** 2) - r)
    return distance

# Example usage:
x1 = 0  # x-coordinate of the point
y1 = 6  # y-coordinate of the point
h = 0   # x-coordinate of the circle center
k = 0   # y-coordinate of the circle center
r = 5   # radius of the circle

distance = shortest_distance_to_circle(x1, y1, h, k, r)
print("Shortest distance between the point and circle:", distance)
```
