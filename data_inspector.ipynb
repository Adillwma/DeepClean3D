```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Define the folder path where the .npy files are stored
folder_path = r"N:\Yr 3 Project Datasets\Dataset 20_X500\Data"

# Get a list of all the .npy files in the folder
file_list = [f for f in os.listdir(folder_path) if f.endswith('.npy')]
                  
# Randomly select a file from the list
file_index = np.random.randint(0, len(file_list))

# Load the file
image = np.load(os.path.join(folder_path, file_list[file_index]))

# Print details about the files properties
print(f"File name: {file_list[file_index]}")
print(f"Shape: {image.shape}")
print(f"Min: {image.min()}")
print(f"Max: {image.max()}")
    
# Plot the image
plt.imshow(image)
plt.show()
```
