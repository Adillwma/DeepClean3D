```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
true_values = np.random.randint(0, 100, size=100)
predicted_values = true_values + np.random.normal(0, 10, size=100)

# Calculate the MSE and MAE for different differences in values
diffs = np.arange(0, 50, 1)
mse_values = []
mae_values = []
for diff in diffs:
    mse = np.mean((true_values + diff - predicted_values)**2)
    mae = np.mean(np.abs(true_values + diff - predicted_values))
    mse_values.append(mse)
    mae_values.append(mae)

# Plot the MSE and MAE values
plt.plot(diffs, mse_values, label='MSE')
plt.plot(diffs, mae_values, label='MAE')
plt.xlabel('Difference in values')
plt.ylabel('Error')
plt.title('MSE and MAE vs Difference in values')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
true_values = np.random.randint(0, 100, size=100)/10000
predicted_values = true_values + np.random.normal(0, 10, size=100)/10000

# Calculate the MSE, MAE, BCE, and SSE for different differences in values
diffs = np.arange(0.2, 1, 0.001)
mse_values = []
mae_values = []
bce_values = []
sse_values = []
for diff in diffs:
    mse = np.mean((true_values + diff - predicted_values)**2)
    mae = np.mean(np.abs(true_values + diff - predicted_values))
    bce = -(true_values + diff) * np.log(predicted_values + 1e-1) - (1 - true_values - diff) * np.log(1 - predicted_values + 1e-1)  
    sse = np.sum((true_values + diff - predicted_values)**2)
    mse_values.append(mse)
    mae_values.append(mae)
    bce_values.append(np.mean(bce))
    sse_values.append(sse)

# Plot the MSE, MAE, BCE, and SSE values
plt.plot(diffs, mse_values, label='MSE')
plt.plot(diffs, mae_values, label='MAE')
plt.plot(diffs, bce_values, label='BCE')
#plt.plot(diffs, sse_values, label='SSE')
plt.xlabel('Difference in values')
plt.ylabel('Error')
plt.title('Loss functions vs Difference in values')
plt.legend()
plt.show()
```
