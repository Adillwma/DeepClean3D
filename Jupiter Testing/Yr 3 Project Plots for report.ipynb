```python
### PLotting ReLu Function output

import numpy as np
import matplotlib.pyplot as plt

# Define the ReLU function
def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))


# Create an array of values from -5 to 5
x = np.linspace(-5, 5, 100)

# Compute the output of the ReLU function for each value of x
y = relu(x)

y_sig = sigmoid(x)


# Plot the ReLU function
plt.plot(x, y, label="ReLu")
plt.plot(x, y_sig, label="Sigmoid")
plt.xlabel('Input')
plt.ylabel('Output')
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.grid(alpha=0.2)

# Add dashed lines at x=0 and y=0
plt.axhline(0, linestyle='--', color='gray')
plt.axvline(0, linestyle='--', color='gray')
plt.legend()
plt.show()
```
