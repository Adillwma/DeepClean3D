```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the radar plot
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
data = [4, 5, 2, 7, 6]  # Values for each category

def create_radar_plot(categories, data):
    # Create a radar plot
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'polar': True})

    # Number of categories
    num_categories = len(categories)

    # Calculate angles for each category
    angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False)

    # Close the plot by repeating the first angle
    angles = np.concatenate((angles, [angles[0]]))

    # Normalize data and close the plot
    data = np.concatenate((data, [data[0]]))
    data_norm = data / max(data)

    ax.plot(angles, data_norm, color='b', marker='o')
    ax.fill(angles, data_norm, color='b', alpha=0.25)

    # Set category labels
    ax.set_xticks(angles[:-1])  # Remove the last angle
    ax.set_xticklabels(categories)
    ax.yaxis.grid(True)

    # Add a title
    plt.title('Radar Plot Example')

    # Show the plot
    plt.show()

create_radar_plot(categories, data)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the radar plot
categories = ['SNR', 'PSNR', 'SSIM', 'NMI', 'CC']
data = np.array(([3, 1, 74, 1, 1], [6, 52, 2, 76, 6], [-4, 272, 2, 1, 6]))  # Values for each category
multi_labels = ['Model 1', 'Model 2', 'Model 3']


def create_radar_plot_multi_model(categories, multi_data, multi_labels):
    # Create a radar plot
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'polar': True})

    # Number of categories
    num_categories = len(categories)

    # Calculate angles for each category
    angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False)

    # Close the plot by repeating the first angle
    angles = np.concatenate((angles, [angles[0]]))

    min_vals = np.min(multi_data, axis=0)
    max_vals = np.max(multi_data, axis=0)

    #normalized_data = (data - min_vals) / (max_vals - min_vals)  # 0-1 NORM
    normalized_multi_data = 0.1 + 0.9 * (multi_data - min_vals) / (max_vals - min_vals) # 0.1-1.0 NORM

    for data_norm, data_label in zip(normalized_multi_data, multi_labels):
        # Normalize data and close the plot
        data_norm = np.concatenate((data_norm, [data_norm[0]]))

        ax.plot(angles, data_norm, marker='o', label=data_label)
        ax.fill(angles, data_norm, alpha=0.25)

    # Set category labels
    ax.set_xticks(angles[:-1])  # Remove the last angle
    ax.set_xticklabels(categories)
    #ax.yaxis.grid(True)
    ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
    # Add a title
    plt.title('Normalised Performance of Tested Models')

    # Show the plot
    plt.show()

create_radar_plot_multi_model(categories, data, multi_labels)
```
