```python
import torch
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def plot_array(arr, title=""):
    if arr.ndim == 4: 
        # remove the first two dimensions (batch and channel) using slicing and reshape
        arr = arr.detach().squeeze(0).squeeze(0).float()
    fig, ax = plt.subplots()


     # set cmap
    cmap = plt.cm.viridis
    ax.imshow(arr, cmap=cmap, vmax=(np.max(arr) * 1.5))


    
    # Add the numerical values as text annotations
    for i in range(arr.shape[0]):
        for j in range(arr.shape[1]):
            ax.text(j, i, arr[i, j], ha="center", va="center", color="w", fontsize=18)

    # Add a colorbar to the plot
    #cbar = ax.figure.colorbar(im, ax=ax)
    fig.suptitle(title)
    # Show the plot
    ax.axis("off")
    plt.show()
```

```python
in_array = np.array(([0,0,0,10],
                    [0,0,40,0],
                    [0,70,0,0], 
                    [100,0,0,0]))

noised_array = np.array(([0,0,0,10],
                    [20,0,40,0],
                    [0,70,0,0], 
                    [100,0,19,35]))

network_out_array = np.array(([0,0,0,5],
                    [0,16,23,0],
                    [0,47,0,0], 
                    [88,0,0,0]))

mask = np.array(([0,0,0,1],
                    [0,1,1,0],
                    [0,1,0,0], 
                    [1,0,0,0]))

final_array  = np.array(([0,0,0,10],
                    [0,0,40,0],
                    [0,70,0,0], 
                    [100,0,0,0]))
```

```python
plot_array(in_array, "Clean Input")
plot_array(noised_array, "Noised Img")
plot_array(network_out_array, "Network Output")
plot_array(mask, "Mask")
plot_array(final_array, "Final Masking Result")
```
