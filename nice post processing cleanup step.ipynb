```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread

image_path = r'A:\Users\Ada\GitHub\DeepClean_Repo\Parabola2 - Epoch 226.png'  # Replace with the actual path to your image
image = imread(image_path, as_gray=True)



plt.imshow(image)
plt.title('Original Image')
plt.xlabel('Pixel')
plt.show()

# plot the square of the abssolute valu eof the imgae
plt.imshow(np.abs(image)**4)
plt.title('Absolute Value Cubed of Image')
plt.xlabel('Pixel')
plt.show()
```

```python

# calulate intesity of image over x
intensity = np.sum(image, axis=1)
plt.plot(intensity)
plt.title('Intensity of Image')
plt.xlabel('Row')
plt.show()



# calulate intesity of image over x
intensity = np.sum(np.abs(image)**4, axis=1)
plt.plot(intensity)
plt.title('Intensity of Image')
plt.xlabel('Row')
plt.show()


# calulate intesity of image over y
intensity = np.sum(image, axis=0)
plt.plot(intensity)
plt.title('Intensity of Image')
plt.xlabel('Column')
plt.show()

# calulate intesity of image over y
intensity = np.sum(np.abs(image)**4, axis=0)
plt.plot(intensity)
plt.title('Intensity of Image')
plt.xlabel('Column')
plt.show()

```
