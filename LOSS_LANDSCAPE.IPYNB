```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the custom loss function
def custom_loss(input, target):
    return (input - target) ** 2

# Define the range and the number of points in the grid
input_range = np.linspace(-10, 10, 100)
target_range = np.linspace(-10, 10, 100)

# Create a mesh grid
input_grid, target_grid = np.meshgrid(input_range, target_range)

# Convert the grid to PyTorch tensors
input_grid_tensor = torch.tensor(input_grid, dtype=torch.float32)
target_grid_tensor = torch.tensor(target_grid, dtype=torch.float32)

# Ensure gradients are enabled
input_grid_tensor.requires_grad_(True)

# Compute the loss values
loss_values = custom_loss(input_grid_tensor, target_grid_tensor)

# Sum the loss values to get a scalar output
loss_scalar = loss_values.sum()

# Compute the gradients with respect to the inputs
loss_scalar.backward()

# Get the gradients
gradients = input_grid_tensor.grad

# Convert tensors back to numpy for visualization
loss_values_np = loss_values.detach().numpy()
gradients_np = gradients.detach().numpy()

# Create a surface plot for the loss landscape
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(input_grid, target_grid, loss_values_np, cmap='viridis')

ax.set_xlabel('Input')
ax.set_ylabel('Target')
ax.set_zlabel('Loss')
plt.title('Loss Landscape')
plt.tight_layout()
plt.show()


plt.figure()
plt.quiver(input_grid, target_grid, gradients_np, gradients_np, cmap='viridis')
plt.show()
```
